# Automated File Sorter in File Explorer - Python Project

### Importing os(Operating systems) & shutil (shell utility) packages

```python
import os, shutil
```

### Viewing files inside TEMP FOLDER

```python
#assigning a variable (Folder_Path) to our current working folder
Temp_Folder_Path = r"C:/Users/sasikanthchowdhary.n/OneDrive - Mu Sigma Business Solutions Pvt. Ltd/Desktop/TEMP FOLDER"

#listdir() opening a folder and seeing what's inside.
Folder_View =  os.listdir(Temp_Folder_Path)
print(Folder_View)
```

### Creating a Organized Folder

```python
#Creating Organized Folder Path
Organized_Folder_Path = os.path.join(Temp_Folder_Path, 'Organized_Folder')


#Checking if Organized_Folder previously exists or not in TEMP FOLDER 
if not os.path.exists(Organized_Folder_Path):
    os.mkdir(Organized_Folder_Path)
    print("Organized_Folder created successfully!")
```

### Creating Sub Folders inside Organzied Folder based on File Types

```python
#Dictionary of Mapping of File Type & Folder Name
FileType_Folder_Map = {
    '.bmp'  : 'bmp_Files',
    '.xlsx' : 'xlsx_Files',
    ".txt"  : 'txt_Files'     
}

#Creating Sub Folders
for Sub_Folder in FileType_Folder_Map.values():
    Sub_Folder_Path = os.path.join(Organized_Folder_Path, Sub_Folder)
    if not os.path.exists(Sub_Folder_Path):
        os.mkdir(Sub_Folder_Path)
        print(f'{Sub_Folder} created successfully!')
```

### Moving files into related Sub Folders Based on file type


```python
for item in Folder_View:
    for File_Type, Folder_Name in FileType_Folder_Map.items():
        if item.endswith(File_Type):
            source = os.path.join(Temp_Folder_Path, item)
            destination = os.path.join(Organized_Folder_Path, Folder_Name, item)
            shutil.move(source, destination)
            print(f'Successfuly moved {item} into {Folder_Name}! ')
```
